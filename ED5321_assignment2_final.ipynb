```python
#Question 2A
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Given constants
mu = 5.515e-4  # linear mass density (kg/m)
T0 = 52.5      # initial tension in open string (N)

# Frequencies (Hz) and corresponding speaking lengths (m)
frequencies = np.array([
    196, 207.650, 220, 233.080, 246.94, 261.62, 277.180,
    311.12, 329.62, 349.22, 369.99, 391.99, 415.30, 466.16,
    493.88,523.25,554.36,622.25,698.45,739.98
])

lfi_m = np.array([
    0.7840, 0.7380, 0.699, 0.662, 0.623, 0.589, 0.556,
    0.496, 0.465, 0.441,0.417,0.396,0.376,0.342, 0.318, 0.304, 0.285, 0.256,
    0.2410, 0.209
])

lsi_m =np.sqrt(lfi_m**2+0.015**2)
# Compute Tsi using Mersennes Law: T = 4 * mu * l^2 * f^2
Tsi = 4 * mu * (lsi_m ** 2) * (frequencies ** 2)

# Compute percentage increase in tension
percent_increase = (Tsi - T0) / T0 * 100

# Create DataFrame for easy inspection
df = pd.DataFrame({
    "f_i (Hz)": frequencies,
    "l_fi (m)": lfi_m,
    "l_si (m)": lsi_m,
     "T_si (N)": Tsi,
    "% Increase in Tension": percent_increase
})

# Sort by frequency
df_sorted = df.sort_values(by="f_i (Hz)").reset_index(drop=True)

# Display the table
print(df_sorted)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df_sorted["f_i (Hz)"], df_sorted["% Increase in Tension"], marker='o', linestyle='-')
plt.title("Percentage Increase in Tension vs Frequency")
plt.xlabel("Frequency at ith fret(Hz)")
plt.ylabel("Percentage Increase in Tension ((Tsi - T0) / T0 * 100%)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Demerits and Merits for the following assumption:
A. Tpi= T0:
Merits:
1.As base tension is known to us it is easy to calculate Tpi values.
2.Loosely speaking nearly a good approximation as change in length of the string is minimal.
3.I think it is better for the instruments where hm is almost equal to hf
Demerits:
1.Physics may change(Geometry) as we press the string.
2.Ends of instruments have variable tension,this assumption may overlook it.

B.Tpi=Tsi
Merits:
1.More reliable as tension is equally distributed on both sides(approximation)
2.Makes good calcultion if hf is high or fret is far from ends.
Demerits:
1.Requires Tsi for each value of i which makes it complex for calculation.
2.Equal tension is not accurate at ends.

Instruments like sitar where hm is more assumption 2 is preferable.Where flat fretted instruments like guiar assumption 1 can be considered.

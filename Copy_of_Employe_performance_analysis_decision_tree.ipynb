```python
#label encoder helps to convert categorical to numerical
enc=LabelEncoder()
for i in (0,2,3,4,5,6,7,16,26):
  data.iloc[:,i] = enc.fit_transform(data.iloc[:,i])
data.head()
```

```python
y=data.PerformanceRating
x=data.iloc[:,[4,5,9,16,20,21,22,23,24]]
x.head()
#data.corr should be studied
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
#decision tree with gridsearchcv

#training the model
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
classifier_dtg=DecisionTreeClassifier(random_state=42,splitter='best')
parameters=[{'min_samples_split':[2,3,4,5],'criterion':['gini']},{'min_samples_split':[2,3,4,5],'criterion':['entropy']}]
model_gridtree=GridSearchCV(estimator=classifier_dtg,param_grid=parameters,scoring='accuracy',cv=10)
model_gridtree.fit(x_train,y_train)#cv is cross verification
```

```python
model_gridtree.best_params_
```

```python
y_predict=model_gridtree.predict(x_test)
```

```python
#random forest with gridsearch cv random forest takes multiple decision trees

#training
from sklearn.ensemble import RandomForestClassifier
classifier_rfg=RandomForestClassifier(random_state=42,n_estimators=23)
parameters=[{'min_samples_split':[2,3,4,5],'criterion':['gini','entropy'],'min_samples_leaf':[1,2,3]}]
model_gridforest=GridSearchCV(estimator=classifier_rfg,param_grid=parameters,scoring='accuracy',cv=10)
model_gridforest.fit(x_train,y_train)
```

```python
model_gridforest.best_params_
```

```python
y_predict=model_gridforest.predict(x_test)
```

```python
print(accuracy_score(y_test,y_predict))
```
